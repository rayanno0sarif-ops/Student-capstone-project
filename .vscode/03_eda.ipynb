```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt     
#import the libraries needed for this task

df = pd.read_csv(r"C:\Users\rayan\OneDrive\Documents\GitHub\Student-capstone-project\data\cleaned\clean data features.csv")     
#load the new data with features
```

```python
#1 histograms / KDE
#i will load 3 numerical histograms and see the shape
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
sns.histplot(df["Age"], color="crimson", kde=True)          #the age distribution has shown to be slightly right skewed so there are more young peopl than old 
plt.title("age")

plt.subplot(1,3,2)
sns.histplot(df["Fare_log"],color="cyan", kde=True)         #the fare log distribution is right skewed with the most common fare being 2 by a long shot
plt.title("fare (log)")

plt.subplot(1,3,3)
sns.histplot(df["FamilySize"],color="teal", kde=True)       #the family size distribution is heavily right skewed with the most commomn family size is 1 also by a long shot
plt.title("family size")

plt.tight_layout()
plt.show()
```

```python
#2 grouped boxplots

plt.figure(figsize=(6,4))
sns.boxplot(x=df["Pclass_ord"], y=df["Survived"], color="beige")      #this compares the survival of passengers based on their seating class 
plt.title("survival by Class")



male_survival = (df["Sex_male"] and df["Survived"]).sum()

plt.figure(figsize=(6,4))
sns.boxplot(x=df["Sex_male"], y=df["Survived"], color ="navy")        #this compares the survival of passengers based on the gender to see wich gender had more survivors
plt.title("Survival by Sex")



plt.show()
```

```python
#3 correlation heatmap

#find the top 10 features most related to survival by using head 10 and not ascending
#higher correlation means the feature affects survival more 
corr = df.corr(numeric_only=True)
top = corr["Survived"].abs().sort_values(ascending=False).head(10)

#this heatmap shows how the top features i got relate to eachother
plt.figure(figsize=(8,6))
sns.heatmap(df[top.index].corr(), annot=True, cmap="viridis")       #i asked google for a different color pallet to look unique
plt.title("Top Correlated Features")
plt.show()


# In this heatmap:
# - Dark yellow/bright colors = strong positive relationship (two features go up together)
# - Dark purple/blue colors = strong negative relationship (one goes up while the other goes down)
# - Middle green colors = weak or no relationship
# The stronger the color, the stronger the connection between the two features.
#the information above is from chatgpbt describing the viridis colors and what they mean
#it also mentioned that colorblind people can see this clearly
```

```python
#4 scatterplot

plt.figure(figsize=(6,4))

sns.scatterplot(x=df["Age"],y=df["Fare_log"],hue=df["Survived"], s=15)    #hue is the third variable, hue is the third variable, it changes the color of each point based on wether they survived or not
#i searched on how to make the dots smaller and found s=int controls the size

plt.title("age and fare with survival")
plt.show()
```

```python
#5 groupby summary

#this shows the average survival rate for each age group when i added that column using bins
#the the higher number means more people in that group survived

group_survival = df.groupby("AgeGroup")["Survived"].mean()  
group_survival
```
