```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt     
from sklearn.preprocessing import StandardScaler
#import the libraries needed for this task

#get the file
df = pd.read_csv(r"C:\Users\rayan\Downloads\capstone project\data\cleaned\clean data features.csv")   
```

```python
#1 compute the mean and standard deviation

#get the column survived as an array
y = df["Survived"].values

#get the mean and std using numpy
mean_survived = np.mean(y)
std_survived = np.std(y)

print(mean_survived)
print(std_survived)
```

```python
#2 standardise a column using broadcasting

#get the age column as an array
X = df["Age"].values

#get mean and std of age using numpy
age_mean = np.mean(X)
age_std = np.std(X)


z = (X - age_mean) / age_std

#print the first 20 standardised values by slicing
print(z[:20])


#make the scaler function easier to use
scaler = StandardScaler()

z_sklearn = scaler.fit_transform(df[["Age"]]).flatten()

#print the first 20 values from sklearn
print(z_sklearn[:20])
```

```python
#3 compute cosine similarity

#some numerical features to compare
features = ["Age", "Fare_log", "FamilySize"]

#turn into matrix
X = df[features].values

#find the index of the highest and lowest Fare_log value
highest_index = df["Fare_log"].idxmax()
min_index = df["Fare_log"].idxmin()

v1 = X[highest_index]
v2 = X[min_index]


# Compute cosine similarity manually, chat gpt helped me
dot_product = np.dot(v1, v2)
v1_mag = np.sqrt(np.sum(v1 ** 2))
v2_mag = np.sqrt(np.sum(v2 ** 2))   
cos_sim = dot_product / (v1_mag * v2_mag)
print(cos_sim)
```

```python
#4 Estimate a probability

#only include passengers with high Fare_log
high = df[df["Fare_log"] > 3]

#the probability that these passengers survived
prob_survival_high_fare = high["Survived"].mean()

print(prob_survival_high_fare)
```
