```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt     
from sklearn.preprocessing import StandardScaler

#import libraries and load my clean file

df_original = pd.DataFrame(pd.read_csv(r"C:\Users\rayan\OneDrive\Documents\GitHub\Student-capstone-project\data\cleaned\clean data.csv"))       

df = df_original.copy()         
```

```python
#1 one hot encoding

#i turn text columns like male and female into numbers
#drop_first=True avoids duplicate columns   (according to google)
df = pd.get_dummies(df, columns=["Sex", "Embarked"], drop_first=True)       #chatgpt told me to use these two

```

```python
#2 ordinal encode

#map numbers to show the order and difference in classes(low to high) 
df["Pclass_ord"] = df["Pclass"].map({1: 3, 2: 2, 3: 1})         
```

```python
#i asked chat gpt what should i scale and it said Age and Fare so they are on a similar range because it helps models understand them better
#3 scaling 2 numerical columns

scaler = StandardScaler()
df[["Age_scaled", "Fare_scaled"]] = scaler.fit_transform(df[["Age", "Fare"]])       
```

```python
#4 domain feature

#i created a family size variable that mesures the sibblings and the parents and the person itself
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1        

#fare per person, chat gpt said this has flaws because it might divide by zero so i added try
try:
    df["Fare_per_person"] = df["Fare"] / df["FamilySize"]           
except ZeroDivisionError:
    df["Fare_per_person"] = df["Fare"] 

```

```python
#5 interaction feature

#combine age and pclass to see how they work together
df["Age_x_Pclass"] = df["Age"] * df["Pclass"]               

```

```python
#6 log transform

sns.histplot(df["Fare"],bins = 40, kde=True, color="red")        #hisplot before applying log
plt.title("Fare before log")
plt.show()

#fare has huge values so i use log to make it more simple       according to chatgpt i should transform Fare
df["Fare_log"] = np.log1p(df["Fare"])

sns.histplot(df["Fare_log"], bins = 40, kde=True, color="navy")        #hisplot after applying log
plt.title("Fare after log")
plt.show()
```

```python
#7 binning

#i will put ages into groups like child, teen, adult
bins = [0, 12, 18, 40, 60, 100] 
labels = ["Child", "Teen", "Adult", "Middle", "Senior"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)


```

```python
#corr>0.95


#get correlation matrix
corr = df.corr(numeric_only=True)


high_corr = corr.unstack().sort_values(ascending=False)


high_corr = high_corr[(high_corr < 1) & (high_corr > 0.95)]

print("corr > 0.95\n\n")
print(high_corr)
```

```python
#9 remove redundant features

#Pclass_ord is better than Pclass, so we drop Pclass
df = df.drop(columns=["Pclass"])
#Fare_log is better than Fare, so we drop Fare
df = df.drop(columns=["Fare"])
#make the new csv with the features
df.to_csv(r"C:\Users\rayan\OneDrive\Documents\GitHub\Student-capstone-project\data\cleaned\clean data features.csv", index=False)
```
